```python
import torch
from transformers import AutoModelForCausalLM, BitsAndBytesConfig, AutoTokenizer
from peft import PeftModel
import os
output_dir1 = "/content/drive/MyDrive/Colab Notebooks/Data/"
output_dir1 = os.path.join(output_dir1, "cleaned_final_Chat7B_checkpoint")
```

```python
# Define base model name
repo_id = "meta-llama/Llama-2-7b-chat-hf"
use_ram_optimized_load=False
# Load Base Model
base_model = AutoModelForCausalLM.from_pretrained(
    repo_id,
    device_map='auto',
    trust_remote_code=True,
)
base_model.config.use_cache = False

# Load Lora adapter
model = PeftModel.from_pretrained(base_model, output_dir1)
```

```python
# Setup tokenizer
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat-hf", trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token
```

```python
# Merge the model and adapter
merged_model = model.merge_and_unload()
```

```python
# Push model to Hugging Face
merged_model.push_to_hub("arcpolar/Ubuntu_Llama_Chat_7B")
```

```python
# Push tokenizer to Hugging Face
tokenizer.push_to_hub("arcpolar/Ubuntu_Llama_Chat_7B")
```
